```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# file = "../data/data_major_vote_10.csv"
# file = "..\data\data_mv.csv"data/data_binarization_10.csv
file = "../data/data_major_vote_10.csv"
df = pd.read_csv(file)
df.head()
```

```python
label_columns = list(set(df.columns) - {"text_id", "text", "fold"})
counts = df[label_columns].apply(pd.Series.value_counts)
counts
```

```python
_ = counts.transpose().plot(kind="bar", figsize=(20, 5))
```

```python
counts = df.groupby("fold").sum()[label_columns]/df.groupby("fold").count()[label_columns]
counts.to_csv("../tmp.csv")
```

```python
fig, ax = plt.subplots(figsize=(30,30))
sns.heatmap(counts, annot=True)
```
